```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
Malaria_Est_Cases = pd.read_csv("Malaria_Est_Cases.csv")
Malaria_Est_Deaths = pd.read_csv("Malaria_Est_Deaths.csv")
Malaria_Est_Incidence = pd.read_csv("Malaria_Est_Incidence.csv")

plt.figure(figsize=(10, 6))

sns.lineplot(data=Malaria_Est_Cases, x="YEAR (DISPLAY)", y="Numeric", label="Case", marker="o")
sns.lineplot(data=Malaria_Est_Deaths, x="YEAR (DISPLAY)", y="Numeric", label="Estimated Deaths", marker="o")
sns.lineplot(data=Malaria_Est_Incidence, x="YEAR (DISPLAY)", y="Numeric", label="Estimated Incident", marker="o")

plt.title('Malaria in Vietnam')
plt.xlabel('Year')
plt.ylabel('Case Counts')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Visulization 2
The dataset contains data on under-five mortality rates (per 1000 live births) with various attributes including year, region, and numerical values for the mortality rate along with low and high estimates. Given this information, a compelling visualization might be to plot the under-five mortality rate over time for different regions to observe trends and changes in mortality rates across different parts of the world.

```python
filtered_df = pd.read_csv("Mortality_Under_5_Months.csv")
Malaria_Est_Deaths = pd.read_csv("Malaria_Est_Deaths.csv")
Malaria_Est_Incidence = pd.read_csv("Malaria_Est_Incidence.csv")

agg_df = filtered_df.groupby(['YEAR (DISPLAY)', 'REGION (DISPLAY)'])['Numeric'].mean().reset_index()

pivot_agg_df = agg_df.pivot(index='YEAR (DISPLAY)', columns='REGION (DISPLAY)', values='Numeric')

plt.figure(figsize=(12, 8))
plt.plot(pivot_agg_df, marker='o', linestyle='-')
plt.title('Under-Five Mortality Rate Over Time by Region (Aggregated)')
plt.xlabel('Year')
plt.ylabel('Average Mortality Rate (per 1000 live births)')
plt.legend(pivot_agg_df.columns, title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

# Show plot
plt.show()
```

## Visulization 3
The visualization above shows the aggregated average TB incidence rates over time, separated by region. Each line represents a different region, allowing us to observe how TB incidence rates have changed globally and within specific regions over the years.

This graph can help identify trends, such as decreases or increases in TB incidence rates, and evaluate the impact of public health interventions and TB control programs across different parts of the world.

```python
tb_incidence_df = pd.read_csv("TB_Incidence.csv")

tb_incidence_filtered_df = tb_incidence_df[['YEAR (DISPLAY)', 'REGION (DISPLAY)', 'Numeric']].dropna()

tb_incidence_agg_df = tb_incidence_filtered_df.groupby(['YEAR (DISPLAY)', 'REGION (DISPLAY)'])['Numeric'].mean().reset_index()

tb_incidence_pivot_df = tb_incidence_agg_df.pivot(index='YEAR (DISPLAY)', columns='REGION (DISPLAY)', values='Numeric')

plt.figure(figsize=(12, 8))
plt.plot(tb_incidence_pivot_df, marker='o', linestyle='-')
plt.title('TB Incidence Rates Over Time by Region (Aggregated)')
plt.xlabel('Year')
plt.ylabel('Average Incidence Rate (per 100,000 population)')
plt.legend(tb_incidence_pivot_df.columns, title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
